```python
import os
from os.path import join
import pandas as pd
import numpy as np

DATAPATH = r"C:\Users\anton\Desktop\PROGETTI\EyeGazeRedirection\MPIIGaze"

ANN_PATH = join(DATAPATH,"annotation_subset")
IMAGES_PATH = join(DATAPATH,"Images")
TRAINING_DATASET = r"C:\Users\anton\Desktop\PROGETTI\EyeGazeRedirection\MPIIGaze\training_tre"
```

```python
# Write all_data dataframe
columns = ["img_path","eyes_landmark_pixel","gaze_tp_screen_px","3D_gaze_camera","6D_head_pose","3D_right_eye_center","3D_left_eye_center"]
newdf = []

def split_and_write(stringline):
    stringline = stringline.replace("\n","").split(" ")

    return stringline[:24],stringline[24:26],stringline[26:29],stringline[29:35],stringline[35:38],stringline[38:]


for group in os.listdir(IMAGES_PATH):
    grp_path = join(IMAGES_PATH,group)
    if os.path.isdir(grp_path):
        for day in os.listdir(grp_path):
            grp_day = join(grp_path,day)
            if os.path.isdir(grp_day):
                imagelist = os.listdir(grp_day)
                imagelist.remove("annotation.txt")
                with open(join(grp_day,"annotation.txt"),"r") as file:
                    filer = file.readlines()
                newdf += [["_".join([group,day,imagelist[n]]),*split_and_write(j)] for (n,j) in enumerate(filer)]


newdf = pd.DataFrame(newdf,columns=columns)
newdf.head()


```

```python
path = join(DATAPATH,"total_ann.csv")
newdf.to_csv(path,columns=newdf.columns)
```

```python
from IPython.display import Image,display
import cv2
from math import asin,atan2,degrees,sqrt
EYE_SHAPE = 64

def convert_string_to_float(string_elem):
    angle_string = string_elem.replace(",","").replace("[","").replace("'","").replace("]","").split(" ")
    return [float(k) for k in angle_string]

def compute_theta(x,z):
    return atan2(x,z)

def compute_phi(x,y,z):
    m = sqrt(x**2+y**2+z**2)
    return asin(y/m)

def get_angles(eye_cords,gaze_cord,dx=False):
    
    x,y,z = convert_string_to_float(eye_cords)
    xg,yg,zg = convert_string_to_float(gaze_cord)

    # if dx:
    #     print(f"{'@'*30} right eye cords {'@'*30}")
    # else:
    #     print(f"{'@'*30} left eye cords {'@'*30}")

    # print("eye coords")
    # print(x,y,z)
    # print("graz coords")
    # print(xg,yg,zg)
    theta = compute_theta(x-xg,z-zg)
    phi = compute_phi(z-xg,y-yg,z-zg)

    theta = degrees(theta)
    if dx:
        theta = - theta
    # if dx:
    #     theta = theta - 180
    #     if theta > 180 : theta -= 360
    #     elif theta < - 180 : theta += 360
    phi = degrees(phi) 
    
    return theta, phi 
    return theta - 180 if theta > 0 else theta + 180, phi 

def cut_eyes(img,points,xoff=20,yoff=20):
    points2mod = list(points)
            # Xmin , Ymin, Xmax, Ymax
    rxmin,rymin,rxmax,rymax = [points2mod[0][0]-xoff,points2mod[0][1]-yoff,points2mod[3][0]+xoff,points2mod[3][1]+yoff]
    lxmin,lymin,lxmax,lymax = [points2mod[6][0]-xoff,points2mod[6][1]-yoff,points2mod[9][0]+xoff,points2mod[9][1]+yoff]
    right = img[rymin:rymax,rxmin:rxmax,:]
    left = img[lymin:lymax,lxmin:lxmax,:]

    rimh,rimw = rymax-rymin,rxmax-rxmin

    rleft = (EYE_SHAPE - rimw )// 2
    rright = EYE_SHAPE - rimw -rleft
    rtop = (EYE_SHAPE - rimh )// 2
    rbottom = EYE_SHAPE - rimh - rtop

    limh,limw = lymax-lymin,lxmax-lxmin

    lleft = (EYE_SHAPE - limw )// 2
    lright = EYE_SHAPE - limw -lleft
    ltop = (EYE_SHAPE - limh )// 2
    lbottom = EYE_SHAPE - limh - ltop

    right = cv2.resize(right,(64,64))
    left = cv2.resize(left,(64,64))

    # right = cv2.copyMakeBorder(right, rtop, rbottom, rleft, rright, cv2.BORDER_CONSTANT)
    right = cv2.flip(right,1)
    # left = cv2.copyMakeBorder(left, ltop, lbottom, lleft, lright, cv2.BORDER_CONSTANT)

    return right,left

def compute_and_cut(line):

    imgpath = join(IMAGES_PATH,line.img_path.replace("_","\\"))
    points = line.eyes_landmark_pixel.replace(",","").replace("[","").replace("'","").replace("]","").split(" ")
    point2draw = [(int(points[j]),int(points[j+1])) for j in range(0,len(points),2)]
    img = cv2.imread(imgpath)

    right_theta,right_phi = get_angles(line["3D_right_eye_center"],line["3D_gaze_camera"],dx=True)
    left_theta,left_phi = get_angles(line["3D_left_eye_center"],line["3D_gaze_camera"])
    # print("Theta, Phi")
    # print(right_theta,right_phi)
    # print(left_theta,left_phi)
    # # print()

    towriter = "rt: {:.3f} ° - rf: {:.3f} °".format(right_theta,right_phi)
    towriteL = "lt: {:.3f} °- lf: {:.3f} °".format(left_theta,left_phi)
    cv2.putText(img,towriter,(50,80),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),2)
    cv2.putText(img,towriteL,(700,80),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),2)

    right_img,left_img = cut_eyes(img,point2draw,15,25)

    img[500:564,:64,:] = right_img
    img[500:564,84:148,:] = left_img

    imgname = line.img_path.split("_")
    imgpath = join(TRAINING_DATASET,"images",imgname[0])
    labelpath = join(TRAINING_DATASET,"labels",imgname[0])

    os.makedirs(imgpath,exist_ok=True)
    os.makedirs(labelpath,exist_ok=True)

    rfilename = "right_"+imgname[1]+"_"+imgname[-1]
    rlabelname = rfilename.rsplit(".")[0]+".txt"
    lfilename = "left_"+imgname[1]+"_"+imgname[-1]
    llabelname = lfilename.rsplit(".")[0]+".txt"
    

    cv2.imwrite(join(imgpath,rfilename),right_img)
    with open(join(labelpath,rlabelname),"w") as f:
        text = " ".join([str(right_theta),str(right_phi)])
        f.write(text)

    cv2.imwrite(join(imgpath,lfilename),left_img)
    with open(join(labelpath,llabelname),"w") as f:
        text = " ".join([str(left_theta),str(left_phi)])
        f.write(text)



    #display(Image(data=cv2.imencode('.png', img)[1]),clear=True)
    
```
